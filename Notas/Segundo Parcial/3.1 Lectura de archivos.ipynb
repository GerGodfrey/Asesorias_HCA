````python
ruta  = "discurso.txt"
archivo = open(ruta)
````

````python
texto = archivo.read().replace(",","").split(" ")
````

````python
print(texto)
````

````python
frecuenciaPalab = []

for x in texto:
    frecuenciaPalab.append(texto.count(x))
````

````python
print("Pares\n" + str(list(zip(texto, frecuenciaPalab))))
````

## Escribiendo un archivo

El modo `"w"` dentro de `open` nos permite crear nuevos archivos.
```python
open(nombreArchivo, "w")
```

````python
# Un pequeño ejemplo
with open("Archivos/contador.txt", "w") as f:
    for i in range(100):
        f.write(str(i) + ",")
        
````

````python
archivo2 = open("Archivos/contador.txt")
````

````python
print(archivo2)

texto = archivo2.read()
````

````python
print(texto)
````
