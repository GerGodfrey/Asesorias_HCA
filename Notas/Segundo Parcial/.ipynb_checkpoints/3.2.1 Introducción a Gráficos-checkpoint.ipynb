<h2 style="color:#d62728">Ejercicio</h2>

Un alumno ha decidido contratar uno de dos servicios de streaming. El alumno quiere saber ¿Cuál es la plataforma más conveniente a largo plazo ? (tomando en cuenta solo el precio). 


a. El servicio A ofrece todo el servicio a un precio de  $350$ pesos por mes. ¿Cuánto deberá pagar el alumno si contrata el servicio 24 meses?  

b. El servicio B ofrece una membresia para el primer mes de $350$ pesos y después $300$ pesos por cada mes adicional ¿Cuánto deberá pagar el alumno si contrata el servicio 24 meses?

c. ¿Cuál es el streaming más conveniente(más barato) para el mes 24 ? (responder vía gráficas)

```python
#1 ,2 ,3 ....24 
meses = range(1,24,1)
```

```python
AcumuladoServicioA = [n * 350 for n in meses]
print(AcumuladoServicioA)
```

```python
#0 ,1 ,2 ....23 
AcumuladoServicioB = [(n-1) * 300 + 350 for n in meses]
print(AcumuladoServicioB)
```

```python
plot(meses,AcumuladoServicioA,'orange',meses,AcumuladoServicioB,'blue')
title("Costos por Streaming")
xlabel("meses")
ylabel("precio acumulado")
legend(['Streaming A',"Streaming B"])
grid(True)
text(22,8100,"A>B")
```

```python
print(f"El precio del servicio A es $ {AcumuladoServicioA[22]} el cuál para el mes 24 es mayor que el precio del servicio B que es $ {AcumuladoServicioB[22]}")
print("Para largo plazo conviene más el servicio B")
```
