```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# creamos un DataFrame
df = pd.DataFrame( {'key': ['A','B','C','D','E','F','A','B','C','D','E','F','A','B','C','D','E','F','A','B','C','D','E','F',],
                  'datosA': list(range(1,25,1)),
                  'datosB': np.random.randn(24),
                  'datosC': np.random.randint(0,10,24)
                  }
)
```

```python
df
```
