## Gráfica de histogramas

Es una representación de los datos en forma de barras, donde la superficie de cada barra es proporcional a la frecuencia de los valores representados.

Nos funcionan para tener una "primera" vista de nuestros datos de interes. 

```python
%pylab
%matplotlib inline
```

```python
# Importar libreria
import matplotlib.pyplot as plt
```

En el documento "votacion" tenemos los votos registrados de un municipio para saber si están en contra o a favor del cambio de horario. Extrae los datos y gráfica cuantos están a favor que está representado con un 1 y cuántos están en contra ( representado con un 0 ).

```python
archivo = open("Archivos/votacion.txt","r")
datos = archivo.read().split(",")
print(datos)
```

## Gráfica de histogramas rotadas

Tenemos los datos de los 10 equipos con más ligas ganadas en toda la historia del futbol mexicano. Grafique los siguientes datos.

```python
equipos= ["América","Guadalajara","Cruz Azul","Toluca","León","UNAM","Tigres","Pachuca","Santos Laguna","Monterrey"]
ligasGanadas = [13,12,11,8,8,7,7,6,6,5]
```

```python
#Librería

#plt.style.use('default')
#Estilo
plt.style.use("Solarize_Light2")

#Definir área de trabajo
fig1,ax = plt.subplots()

# Anexar la información y estilo de gráfica
ax.barh(equipos,ligasGanadas)

#Invertir gráfica
#ax.invert_yaxis()

#Etiquetas
ax.set_xlabel("Total")
ax.set_ylabel("Equipos")
ax.set_title("Historial Ligas Ganadas")

#Imprimir
plt.show()
```

```python
#Función mágica
#plt.style.use("default")
plt.style.use("Solarize_Light2")
barh(equipos,ligasGanadas)
title("Historial Ligas ganadas")
xlabel("Total")
ylabel("Equipos")
```
