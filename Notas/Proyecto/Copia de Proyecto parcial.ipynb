### Introduce el nombre del archivo con su extension .csv y debe de estar en la misma carpeta que este archivo El archivo debe de tener 3 columnas (salario,bonos,horasextra) y la cantidad de filas deseadas, sin embargo la primera sera la de los títulos

```python
def salariot (x):
    # Pasamos el archivo a lista
    import csv
    with open(x) as f:
        reader = csv.reader(f)
        datos = list(reader)
    
    gastosporPersona = []
    cont = 0
    suma = 0
    totales = 0 
    for x in range(1,len(datos),1):

        for i in datos[x]:
            totales = totales + 1 
            if(cont < 3):
                cont = cont + 1
                suma = suma + float(i)            
            else:
                gastosporPersona.append(suma)

                suma = 0
                suma = suma + float(i)
                cont = 1

    gastosporPersona.append(suma)
    return gastosporPersona
```

```python
salariot=salariot("InformacionProyecto.csv")
```

```python
def salariosmenosimpuesto (lista_salarios_totales):
    listaconimpuesto=[]
    for x in lista_salarios_totales:
        if x<=40000:
            x=x-(x*.10)
            listaconimpuesto.append(x)
        elif 40000<x and x<=80000:
            x=x-(x*.15)
            listaconimpuesto.append(x)
        else:
            x=x-(x*.20)
            listaconimpuesto.append(x)
            
    return listaconimpuesto
```

```python
salariosmenosimpuesto(salariot)
salariofinal=salariosmenosimpuesto(salariot)
```

```python
def rangosdesaldos (lista_salarios_totales):
    listacontadores=[]
    contador=0
    contador1=0
    contador2=0
    for x in lista_salarios_totales:
        if x<=40000:
            contador=contador+1
            
        elif 40000<x and x<=80000:
            contador1=contador1+1
            
        else:
            contador2=contador2+1
    listacontadores.append(contador)
    listacontadores.append(contador1)
    listacontadores.append(contador2)
    
            
    return listacontadores
    
```

```python
listarangosaldo=rangosdesaldos(salariosmenosimpuesto(salariot))
print(listarangosaldo)
```
