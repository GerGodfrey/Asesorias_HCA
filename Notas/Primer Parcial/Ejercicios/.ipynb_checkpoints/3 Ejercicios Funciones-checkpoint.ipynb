## Ejercicio 1

1) Defina una función llamada 'peri' tal que, dadas las longitudes de los lados A y B de un rectángulo, calcule el perímetro. El perímetro de un rectángulo se calcula:  
2\*(lado A) + 2\*(lado B)

2) Defina una función llamada 'area' tal que, dadas las longitudes de los lados A y B de un rectángulo, calcule el área. El área o superficie de un rectángulo se calcula:  
(lado A) \* (lado B)

3) Calcule el perímetro y el área de un rectángulo de lados 4.5 y 6.5 usando las funciones creadas de los incisos a) y b)

4) Defina una función llamada 'geoT' tal que, dadas las longitudes de los lados A y B de un rectángulo, calcule y entregue una tupla con el perímetro y el área del rectángulo; defina otra función 'geoD' que calcule las mismas cantidades en un diccionario.

```python
#GENÉRICAS
def peri(a,b):
    return (2*a)+(2*b)
    
def area (a,b):
    return a*b

peri =  peri(4.5,6.5)
area = area(4.5,6.5)

print(peri)
print(area)
```

```python
#DICCIONARIO
def geoD(a,b):
    dic = {'peri':(2*a)+(2*b) , 'area': a*b}
    return dic
geod = geoD(4.5,6.5)
print(geod)

```

```python
#TUPLA
def geoT(a,b):
    t=( (2*a)+(2*b) , a*b )
    return t

geo = geoT(4.5,6.5)
print(geo)
```

### ejercicio 2
Escriba una función que a partir de una pareja de números $(a,b)$ calcule la siguiente operación:  
* $\frac{(a-b)^2}{(a-b)}$, si los números son distintos
* 0, si los números son iguales 

Aplique la función creada a diversos casos.

```python
def eje2 (a,b):
    if (a!=b) :
        
        res= (a-b)**2  / (a-b)
    else :
            res=0
            
    return res

s = eje2(2,3)
print(s)
    
```

### ejercicio 3
Una tienda ofrece descuentos a sus clientes de acuerdo a la siguiente política:
- 5%, si el monto de compra es inferior a \$ 800.00;
- 15%, si el monto de compra es mayor o igual a \$ 800.00 y menor a $ 1,500.00;
- 25%, si el monto de compra es mayor o igual a \$ 1,500.00.  


Escriba una función que use como entrada un valor del monto de compra, y calcule como salida el precio pagado por el cliente con el descuento incluído según el criterio dado.

Aplique la función creada a los montos dados en la siguiente lista:
1500.00, 224.50, 525.00, 2000.00, 1250.00, 825.00 


```python
def eje3 (monto):
    if monto < 800:
        des = .05
    elif (800 <= monto and monto < 1500):
            des = .15
    else:
                des = .25
                
    res = monto - (monto*des)
    return res

lista = (1500.00, 224.50, 525.00, 2000.00, 1250.00, 825.00)

respuestas = [ eje3(x) for x in lista]
print(respuestas)
```

```python
def eje8 (contr,cont):
    contraseña = 'PasareHCA'
    res=''
    
    if(cont > 3):
        print("Ya no puedes intentarlo")
    elif (cont <= 3 and contr == contraseña):
        print("Bienvenido")
    elif (cont <= 3 and contr != contraseña):
        print("Intentalo de nuevo")
        cont = cont + 1
    return cont
    
```
