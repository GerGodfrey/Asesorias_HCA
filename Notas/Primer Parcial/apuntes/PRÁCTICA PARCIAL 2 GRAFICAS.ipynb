**EJERCICIO 2**
a) El hotel A ofrece una habitación a un precio de 900 pesos por noche. 
¿Cuánto deberá pagar esa persona si se hospeda 'n' noches? Escribe una expresión para el total como función del número de noches 'n'; grafica esta expresión como una sucesión de puntos. 


```python
def ha(n):
    t=n*900
    return t
```

```python
ha(1)
```

```python
ha(3)
```

```python
n=list(range(1,21,1))
```

```python
# lista de precios por noche hotel A
lna=list(map(ha,n)) #el map es para que pase por todos los valores
print(lna)
```

```python
plot(lna)
title("Precios del Hotel A")
xlabel("noches")
ylabel("precios")
print("")
```

b) El hotel B ofrece una habitación equivalente a la del hotel A en 1,500 pesos la primera noche y 850 pesos por cada una de las noches siguientes. ¿Cuánto deberá pagar el cliente si se hospeda 'n' noches? Escribe una función (def)  cuyo argumento sea el número de noches 'n'.  


```python
def hb(n):
    if n==1:
        t=1500
    else:
        t=1500+850*(n-1)
    return t
```

```python
lnb=list(map(hb,n))
print(lnb)
```

```python
plot(lnb,"r:")
title("Precio Hotel B")
xlabel("Noches")
ylabel("Precio")
print("")
```

c) ¿Cuál es el hotel más conveniente? Responde la pregunta a partir de un examen de las gráficas, sin realizar ningún cálculo. Considera que ambos hoteles ofrecen un servicio equivalente.

```python
plot(lna,"r:",lnb,"b--")
axis([9,20,9000,17500]) #cambiar el eje
print("")
```

```python
i=0
conv=[]
while i<len(lna):
    if lna[i]<lnb[i]:
        co="Hotel A"
    else:
        co="Hotel B"
    conv.append(co)
    i=i+1
print("El hotel que mas conviene es: \n",conv)
```

```python
#Para conocer hasta que noche conviene el hotel B
conv.index("Hotel B")
```

**Ejercicio 3**
Una importante agencia de alquiler de bicicletas cobra, por la renta de una bicicleta, 14 pesos, más 0.15 pesos por km. Una segunda agencia, también importante, cobra 20 pesos, más 0.05 pesos por km. ¿Qué agencia ofrece el mejor trato? Realiza una gráfica que permita tomar la decisión.

```python
def ag1(km):
    t=14+0.15*km
    return t
```

```python
def ag2(km):
    t=20+0.05*km
    return t
```

```python
km=list(range(1,71,1))
```

```python
# recordar que lo tengo que convertir a lista:
lag1=list(map(ag1,km))
lag2=list(map(ag2,km))
print("La lista de precios de la agencia 1 es: \n",lag1)
print("La lista de precios de la agencia 2 es: n",lag2)
```

```python
plot(km,lag1,"r",km,lag2,"b")
title("Agencias de Bicis")
grid(True)
print("")
```

```python
i=0
cambio=[]
while i<len(km):
    if lag1[i]<lag2[i]:
        camb="Agencia 1"
    else:
        camb="Agencia 2"
    cambio.append(camb)
    i=i+1
print("El hotel que más conviene es: \n",cambio)
```

```python
#Para sabe hasta que punto conviene la 2
cambio.index("Agencia 2")
```

```python
lage1=[14+km*0.15 for km in range(1,71,1)]
lage2=[20+km*0.05 for km in range(1,71,1)]
print("Lista de precios agencia 1 es: \n",lage1)
print("Lista de precios agencia 2 es: \n",lage2)
```

```python
plot(lage1,"blue",lage2,"red")
title("Precios Agencias de Bicis")
print("")
```

# Gráfica circular
* etiquetas (labels=lista)
* colores (colors=lista)
* separación (explode=lista)
* auto porcentaje (autopct="%1.2f%%")

```python
nombres="Adaya Isui Andrés Fernanda Jesús Ana"
nom=nombres.split()
edades1=list(randint(18,22,6))
print(nom)
print(edades1)
```

```python
pie(edades1,colors=cole) #cuando es circular es colors
title("Edades amigos") 
print("")
```

```python
pie(edades1,colors=cole,autopct="%2.2f%%") #2 entero, porcentaje 2 decimales, la f y los dos procetajes
print("")
```

```python
sep=[0,.1,0,0,0,0] #al que separo le pongo .1
pie(edades1,colors=cole,autopct="%2.2f%%",explode=sep) #2 entero, 2 decimales, la f y los dos procetajes
print("")
```
